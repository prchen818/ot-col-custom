```python
import pandas as pd

data = pd.read_csv("kafka_metrics.csv")
data.head()
```

```python
import matplotlib.pyplot as plt

# 每隔10条采样一个点
sampled_data = data.iloc[::30].copy()

# 转换时间字段为 datetime 类型
sampled_data['timestamp'] = pd.to_datetime(sampled_data['timestamp'])

# 第一张图：时间线 vs 堆积量
plt.figure(figsize=(12, 5))
plt.plot(sampled_data['timestamp'], sampled_data['lag'], marker='o')
plt.xlabel('时间线')
plt.ylabel('堆积量')
plt.title('堆积量随时间变化折线图（每10条采样）')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 第二张图：时间线 vs 等待时间
plt.figure(figsize=(12, 5))
plt.plot(sampled_data['timestamp'], sampled_data['wait_ms'], marker='o', color='orange')
plt.xlabel('时间线')
plt.ylabel('等待时间 (ms)')
plt.title('等待时间随时间变化折线图（每10条采样）')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
